```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_data = sns.load_dataset('titanic')
print(titanic_data.head())
```

```python
pivot_table = titanic_data.pivot_table(
    index='class',
    columns='sex',
    values='survived',
    aggfunc='mean',
    observed='true'
)
print(pivot_table)

```

```python
count_pivot_table = titanic_data.pivot_table(
    index='class',
    columns='sex',
    values='survived',
    aggfunc='count',
    observed='true'
)
print(count_pivot_table)
```

```python
fare_pivot_table = titanic_data.pivot_table(
    index='class',
    columns='embark_town',
    values='fare',
    aggfunc='mean',
    observed='true'
)
print(fare_pivot_table)
```

```python
fare_pivot_table = titanic_data.pivot_table(
    index=['class', 'sex'],
    columns='embark_town',
    values='fare',
    aggfunc='mean',
    observed=True
)
print(fare_pivot_table)
```

```python
pivot_bar2 = titanic_data.pivot_table(
    values="survived",
    index="sex",
    aggfunc="sum"
)
pivot_bar2.plot(kind="bar", color=["orange", "green"], edgecolor="black", figsize=(6, 6))
plt.title("Survival Count by Gender")
plt.ylabel("Number of Survivors")
plt.xlabel("Gender")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python

plt.figure(figsize=(6, 6))
sns.boxplot(data=titanic_data, x='survived', y='age', hue='survived', palette='Set2', dodge=False)
plt.title("Age Distribution by Survival Status")
plt.ylabel("Age")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.grid(axis='y', linestyle="--", alpha=0.7)
plt.legend([], [], frameon=False)
plt.show()
```
